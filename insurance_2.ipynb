```python

data_test['Id']=data_test.index+1

df_test = data_test.drop("Id", axis =1)

```

```python
columns_with_missing_values = df_test.columns[df_test.isnull().any()]
```

```python
columns_with_missing_values
```

```python
missing_values_info = pd.DataFrame(columns=['Column_Name', 'Missing_Values_Count'])
for column in columns_with_missing_values:
    missing_values_info = missing_values_info.append({
        'Column_Name': column,
        'Missing_Values_Count': df_test[column].isnull().sum()
    }, ignore_index=True)

print(missing_values_info)
```

```python
columns_to_drop = ['Insurance_History_5', 'Family_Hist_2', 'Family_Hist_3',
                   'Family_Hist_5', 'Medical_History_10', 'Medical_History_15',
                   'Medical_History_24', 'Medical_History_32']
df_test = df_test.drop(columns=columns_to_drop)
```

```python
df_test.shape
```

```python
req_cols=['Employment_Info_1','Employment_Info_4','Employment_Info_6','Family_Hist_4','Medical_History_1']
for column in req_cols:
    plt.figure(figsize=(6, 4))
    sns.histplot(df_test[column], kde=True)
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.show()
```

```python
#since they are Distributive Functions(Continuous variable),
#graph is right-skewed,
#we use median
df_test[req_cols]=df_test[req_cols].fillna(df_test[req_cols].median())
df_test.isnull().sum()
```
